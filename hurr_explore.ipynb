# Handling D-types

Converting latitude and longitude to float

```python
def coordinate_mapping(x):
    coord = float(x[:-1])
    if x[-1]=="W":
        coord *= -1
    if x[-1]=="S":
        coord *= -1
    return coord
```

```python
hurr_df["Latitude"] = hurr_df["Latitude"].apply(coordinate_mapping)
hurr_df["Longitude"] = hurr_df["Longitude"].apply(coordinate_mapping)
```

```python
hurr_df.shape
```

Converting Date column to datetime(64)

```python
def dtype(arg):
    hurr_df[arg] = hurr_df[arg].astype('string')
hurr_df['Date'] = pd.to_datetime(hurr_df['Date'])
```

```python
hurr_df['Year'] = hurr_df['Date'].map(lambda x: x.year)
```

```python
hurr_df.info()
```

Convert Maximum Wind to Float

```python
hurr_df['Maximum Wind'] = hurr_df['Maximum Wind'].astype(float)
hurr_df.info()
```

Checking for negative values

```python
negatives =[]
for i in hurr_df['Maximum Wind']:
    if float(i) < 0:
        negatives.append(i)
        
print(negatives)
```

Dropping negaive values

```python
hurr_df = hurr_df.replace('-999', np.nan)
hurr_df = hurr_df.replace('-99', np.nan)
```
